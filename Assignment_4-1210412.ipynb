- Does they have skweness?

A skewed distribution is a distribution that departs from symmetry in the sense that most of the cases are concentrated at one end of the distribution.

    Both the Age and the Fare have positive skewness values, suggesting a right-skewed distribution. This skewness indicates that the data is stretched out towards higher values, with a longer tail on the right side of the distribution.

```python
# Calculate skewness for Age and Fare
age_skewness = df['Age'].skew()
fare_skewness = df['Fare'].skew()

print("Skewness of Age:", age_skewness)
print("Skewness of Fare:", fare_skewness)
```

- Select two categorical features and verify if the data is balanced between the categories of each feature

    I selected both the Sex and Pclass to verify if the data is balanced or not:
<blockquote> I calculated the count for each value within the categories of Sex and Pclass. Then, I verified if these values were numerical using the following equation:  <span style="font-weight:bold">min_value >= (0.5 * max_value)</span> if the result is true so the feature is balance.
    
- The result was as follows:
    - Sex feature is balanced
    - Pclass feature is not balancede.

```python

# Calculate count of each category
sex_counts = df['Sex'].value_counts()
pclass_counts = df['Pclass'].value_counts()

# Compare counts to check for balance
is_sex_balanced = sex_counts.min() >= 0.5 * sex_counts.max()
is_pclass_balanced = pclass_counts.min() >= 0.5 * pclass_counts.max()

print("Is Sex feature balanced?", is_sex_balanced)
print("Is Pclass feature balanced?", is_pclass_balanced)
```

#### This can be seen from the following illustration:

```python
plot.figure(figsize=(10, 4))

plot.subplot(1, 2, 1)
plot.bar(sex_counts.index, sex_counts.values, color='#FFC1C3')
plot.title('Distribution of Sex')
plot.xlabel('Sex')
plot.ylabel('Count')

plot.subplot(1, 2, 2)
plot.bar(pclass_counts.index, pclass_counts.values, color='#D598CF')
plot.title('Distribution of Pclass')
plot.xlabel('Pclass')
plot.ylabel('Count')

plot.tight_layout()
plot.show()
```
